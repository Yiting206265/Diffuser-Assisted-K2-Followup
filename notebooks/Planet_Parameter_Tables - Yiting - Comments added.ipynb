# Planet parameter tables for K2-28b and K2-100b 
- Read in the planet parameters from the MCMC fits and create a table containing the information for K2, Ground, Joint fit for K2 and Joint fit for ground

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append("../src/")

#Read in the K2, Ground, Joint fit for K2 and Joint fit for Ground data for K2-28b and K2-100b
df_ground_k2_28b = pd.read_csv("../data/k2_28b/k2-28b_ground_for_latex_table.csv",sep=",",header=0)
df_k2_k2_28b = pd.read_csv("../data/k2_28b/k2-28b_k2_for_latex_table.csv",sep=",",header=0)
df_jointground_k2_28b=pd.read_csv("../data/k2_28b/k2-28b_jointground_for_latex_table.csv",sep=",",header=0)
df_jointk2_k2_28b=pd.read_csv("../data/k2_28b/k2-28b_jointk2_for_latex_table.csv",sep=",",header=0)

df_ground_k2_100b = pd.read_csv("../data/k2_100b/k2-100b_grounddata_for_latex_table.csv",sep=",",header=0)
df_k2_k2_100b = pd.read_csv("../data/k2_100b/k2-100b_k2data_for_latex_table.csv",sep=",",header=0)
df_jointground_k2_100b=pd.read_csv("../data/k2_100b/k2-100b_jointground_for_latex_table.csv",sep=",",header=0)
df_jointk2_k2_100b=pd.read_csv("../data/k2_100b/k2-100b_jointk2_for_latex_table.csv",sep=",",header=0)
```

```python
#Create planet parameter table for K2-28b
k2_28b_table=pd.DataFrame({'Labels': df_ground_k2_28b['Labels'],
                            'Description':df_ground_k2_28b['Description'],
                            'K2': df_k2_k2_28b['K2'],
                            'Ground':df_ground_k2_28b['Ground'],
                            'Joint fit for Ground':df_jointground_k2_28b['Ground'],
                            'Joint fit for K2':df_jointk2_k2_28b['K2']})
k2_28b_table=k2_28b_table[['Labels','Description','K2','Ground','Joint fit for Ground','Joint fit for K2']]

#Create planet parameter table for K2-100b
k2_100b_table=pd.DataFrame({'Labels': df_ground_k2_100b['Labels'],
                            'Description':df_ground_k2_100b['Description'],
                            'K2': df_k2_k2_100b['K2'],
                            'Ground':df_ground_k2_100b['Ground'],
                            'Joint fit for Ground':df_jointground_k2_100b['Ground'],
                            'Joint fit for K2':df_jointk2_k2_100b['K2']})
k2_100b_table=k2_100b_table[['Labels','Description','K2','Ground','Joint fit for Ground','Joint fit for K2']]
```
